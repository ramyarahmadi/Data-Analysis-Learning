# 1-Creating a DataFrame

```python
import pandas as pd
Data = [[1,7,0],[2,8,3],[3,1,4]]
df = pd.DataFrame(Data,index=['A','B','C'],columns=['x1','x2','x3'])            
#print(df.dtypes)
df
```

# 2-Editing rows and columns:

Assign a new list of values to `df.loc[row_label]` to overwrite every value in that row. The list must match the number of columns, in column order.

```python
df.loc['A']=[1,1,20]
df
```

Updating an Entire Column:

Assign a new list of values directly to `df['column_name']` to overwrite every value in that column. The list must match the number of rows.

```python
df['x3']=[13,17,9]
#df.loc[:,'x3']=[13,17,9]
df
```

`.loc[:, col] = ...` vs. `df[col] = ...`

Both statements look equivalent, but they don't always behave the same way. Direct column assignment (`df['x3'] = [...]`) replaces the column outright, while `.loc[:, 'x3'] = [...]` writes into the existing column in place and is more sensitive to shape or dtype mismatches, which is why it can raise an error where the direct assignment doesn't. When the goal is simply to replace all of a column's values, `df['column'] = [...]` is the safer choice.

```python
df.loc['A','x2']=[1]
df
```

```python
df.loc[['B','C'],['x1','x2']]=[[2,2],[30000,3]]                       #or df.loc[['B','C']['x1','x2']] = 1
df
```

**Adding a New Column**

Assigning to a column name that doesn't exist yet creates it: `df['new_col'] = [...]`.

```python
df['x4']=[11,12,13]
df
```

**Adding a New Row**

Assigning to a row label that doesn't exist yet creates it: `df.loc['new_row'] = [...]`.

```python
df.loc['D']=[400,4,63,17]
df.loc['E']=[5,5,27,11]
df.loc['F']=[6000,6,40,13]
df
```

# 3-Data types

A couple of commands worth remembering when exploring a DataFrame.

**Changing Column Data Types**

Use `.astype('type')` to convert data. Note that `df.astype('float')` on its own returns a **new** DataFrame and does not modify `df` in place; to permanently change a single column, reassign it: `df['col'] = df['col'].astype('float')`.

```python
df.astype('float')    # returns a new float DataFrame, but doesn't change df in place
df['x4'] = df['x4'].astype('float')   # this permanently converts the x4 column
df
```

# 4-Dropping Rows and Columns
The `.drop()` method removes rows or columns. Always pass `inplace=True` (or reassign the result) — otherwise the original DataFrame is left unchanged.

**Dropping Columns**

```python
df.drop(columns=['x1'], inplace=True)   # use inplace=True rather than reassigning to a new variable
df   # x1 was treated as an outlier column, so we removed it from the dataset
```

**Dropping Rows**

```python
df.drop(index=['A'], inplace=True)
df
```

# 5-Rename
**Renaming Columns**

Use `.rename(columns={...})` with a dictionary that maps old names to new names.

```python
df.rename(columns={'x2':'id','x3':'age','x4':'Blood pressure'},inplace=True)   # by dict
df
```

**Renaming Row Labels (Index)**

Use `.rename(index={...})` the same way to relabel rows.

```python
df.rename(index={'B':'Man','D':'Man','F':'Man','C':'Woman','E':'Woman'}, inplace=True)
df
```

# 6-Replacing Values
`.replace()` lets you substitute specific values anywhere in the DataFrame, in a single column, or in a single cell.

```python
#way_1
df.replace({'age':{9:20}},inplace=True) #change item in the columns that you pass to code
df
```

```python
#way_2
df.replace({17:30}, inplace=True)  # replace every occurrence of 17 with 30, anywhere in the DataFrame
df
```

## 7-Working with a Real, Messy Dataset
From here on, we load a CSV file that has some formatting issues and missing values, and clean it up in this lesson and future step by step.

**7.1-Dropping Unneeded Columns**

Remove the unnamed index column (`'0'`) and the `'urban'` column, which aren't needed for this analysis.

```python
df_co.drop(columns={'0','urban'},inplace=True)
df_co
```

**7.2-Setting a meaningful index**

Replace the default numeric index with the `'state'` column so each row is labeled by state name.

```python
df_co.rename(index=df_co['state'],inplace=True)
df_co
```

**Dropping the Original 'state' Column**

Now that its values are being used as the index, the original `'state'` column can be dropped.

```python
df_co.drop(columns={'state'},inplace=True)
df_co
```

**7.3-Shortening Column Names**

Rename the columns to shorter, more convenient labels.

```python
df_co.rename(columns={'Cases':'ca','population':'pop','gdp':'g'},inplace=True)
df_co
```

# 8-Converting Object Columns to Numeric

**Why This Matters**

Columns imported from a CSV are sometimes read as `object` (text) even when they contain numbers, usually because of stray characters or inconsistent formatting. Numeric columns should always be converted to a proper numeric dtype before doing any calculations. `pd.to_numeric()` handles this conversion, and the `errors='coerce'` argument turns any value that can't be parsed into `NaN` instead of raising an error.

```python
df_co.info()
```

```python
# if we see number in data and type of them is object we can this way pd.to_numeric(df_co['cases'],errors='coerce')
```

```python
?pd.to_numeric
```

```python
df_co['cases']  = pd.to_numeric(df_co['cases'],errors='coerce')
# note: the dtype is still 'object' at this point because only one column has been converted so far
```

```python
df_co['cases']  = pd.to_numeric(df_co['cases'],errors='coerce')
df_co['pop'] = pd.to_numeric(df_co['pop'],errors='coerce')
df_co['g']   = pd.to_numeric(df_co['g'],errors='coerce')
df_co
```

# 9-Handling Missing Data

**9.1-Detecting Missing Values**

Use `.isna()` to see a boolean mask of missing values, or `.isna().sum()` to get a per-column count — much more practical on larger datasets.

```python
df_co.isna()            # isn't good when we have a lot of data but another one by .sum() is nececessary.
#or
df_co.isna().sum()      
```

**9.2-Removing Missing Values**

`.dropna()` removes every row that contains at least one `NaN` value.

```python
df_co.dropna(inplace=True)    #remove all row that have NaN value in thier coulmns.
df_co
```

**9.3-Filling Missing Values**

`.fillna()` replaces `NaN` values with a value of your choice — either a single value applied to the whole DataFrame, or a dictionary specifying a different fill value per column.

```python
df_co.fillna(0,inplace=True)    # fill all row that have NaN value by 0 value.
#or
df_co.fillna({'cases':0,'pop':1,'g':-1},inplace=True)  # fill all row that have NaN value by yourselves value.
```
