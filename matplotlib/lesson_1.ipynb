# 📊 Data Visualization Guide in Python

In this guide, we explore essential visualization techniques for Machine Learning and Data Analysis.
# 🗺️ Roadmap Overview

**1. Matplotlib: The Foundation**

Matplotlib is the core plotting library. While low-level, it provides complete control over every aspect of a figure.

**1.1 Line Plot () :** Used for trends and time-series data (e.g., training loss/accuracy over epochs).
                                      
        Key Parameters: x, y (data), color, linestyle, marker, label.

Example: `plt.plot(epochs, loss, color='blue', linestyle='--', marker='o', label='Training Loss')`

**1.2 Pie Chart () :** Used for displaying categorical composition.

        Key Parameters: x (values), labels, autopct (formatting percentages), explode (highlighting a slice).

Example: `plt.pie(counts, labels=categories, autopct='%1.1f%%', explode=[0.1, 0, 0])`

**2. Seaborn: High-Level Statistical Plots**

Seaborn is built on top of Matplotlib and integrates seamlessly with Pandas DataFrames.

**2.1 Swarm Plot () :** A categorical scatter plot where points are adjusted to avoid overlap. Ideal for showing the distribution of data points in small datasets.

        Key Parameters: data, x, y, hue, size. (hue: Semantic mapping (color grouping). size: Radius of the markers.)

Example:`sns.swarmplot(data=df, x='category', y='value', hue='group', size=4)`

**2.2 Box Plot () :** Used for visualizing data distribution and detecting outliers through quartiles.

        Key Parameters: data, x, y, showfliers (boolean, to hide/show outliers).
        
Example:`sns.boxplot(data=df, x='feature_a', y='target')`

**2.3 Strip Plot () :** A strip plot is a categorical scatter plot that displays individual data points along a single axis to visualize the distribution and density of observations.

Example:`sns.stripplot(data=df, x='Category', y='Value', hue='Group', jitter=True)`

**3. Best Practices for Visualization in ML**

**Consistency:** Use the same style/theme for your thesis figures. **Labeling:** Always include xlabel, ylabel, and a legend for clarity. **Resolution:** Use plt.savefig('filename.png', dpi=300) for high-quality images in your academic documents. **Integration:** Always convert processed data (e.g., after SimpleImputer) back to a Pandas DataFrame to leverage Seaborn’s column-based plotting easily.

````python
import matplotlib.pyplot as plt
import numpy as np

# Data Generation
x = np.linspace(0, 10, 100)
y = np.exp(-x/3) * np.cos(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, 
         color='teal',       # color: Hex, RGB, or name
         linestyle='--',     # linestyle: '-', '--', '-.', ':', 'None'
         linewidth=2,        # linewidth: thickness of the line
         marker='o',         # marker: 'o', 's', '^', 'x', etc.
         markersize=4,        # markersize: size of the marker points
         label='Signal Decay')# label: used for the legend

plt.title("Signal Decay Over Time", fontsize=14)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend(loc='upper right') # loc: position of the legend
plt.grid(True, alpha=0.3)    # grid: adds background grid with transparency
plt.show()
````

# 3. Best Practices for Visualization in ML: 
**When visualizing experimental results, model performance, or data preprocessing pipelines, following these standards ensures your work is publication-ready and reproducible**

**1. Visual Consistency**Principle: Maintain a uniform aesthetic across all figures in your thesis or paper. Use consistent color palettes, font sizes, and line weights.

`Pro-tip: Define a global style using plt.style.use('seaborn-v0_8-paper') or plt.rcParams at the start of your script.`

**2. Explicit Labeling & Legend Clarity** 

**Principle:** A plot without context is useless. Every figure must include:

`Axis Labels: With units where applicable (e.g., “Epochs”, “Loss (LL)”, “Learning Rate (ηη)”).`

`Legends: To distinguish between different models, hyperparameter sets, or data splits.`

`Titles: A descriptive title that summarizes the finding.`

**3. High-Resolution Output (Publication Quality)** 

**Principle:** Standard screen resolutions are insufficient for printed documents. Always export figures using a high DPI (Dots Per Inch) to prevent pixelation in your thesis.

`Command: plt.savefig('figure_name.png', dpi=300, bbox_inches='tight')`

`Note: bbox_inches='tight' ensures that labels are not cut off during saving.`

**4. Seamless Data Integration (Pipeline Continuity)** 

**Principle:** Scikit-learn transformers (like SimpleImputer or StandardScaler) often return NumPy arrays, which strip away column names. To use advanced Seaborn features like hue or style, always wrap your processed data back into a Pandas DataFrame.

# 💻 Comprehensive Implementation Example
The following code demonstrate a complete workflow: handling missing data, converting it back to a DataFrame, applying consistent styling, and saving a high-resolution plot.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer

# 1. Setup: Consistency (Global Styling)
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({'font.size': 12, 'font.family': 'serif'})

# 2. Data Preparation: Integration (Scikit-learn -> Pandas)
raw_data = pd.DataFrame({
    'Agent_Type': ['DRL', 'DRL', 'Supervised', 'Supervised', 'Imputation', 'Imputation'],
    'Reward': [10.5, np.nan, 8.2, 7.9, 9.1, 8.8], # Contains NaN
    'Stability': [0.9, 0.8, 0.6, 0.5, 0.7, 0.75]
})

# Preprocessing with SimpleImputer
imputer = SimpleImputer(strategy='mean')
# We only impute the numerical columns
imputed_values = imputer.fit_transform(raw_data[['Reward', 'Stability']])

# CRITICAL STEP: Convert back to DataFrame to preserve column names for Seaborn
df_processed = pd.DataFrame(imputed_values, columns=['Reward', 'Stability'])
df_processed['Agent_Type'] = raw_data['Agent_Type']

# 3. Visualization: Labeling & Legend
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_processed, 
    x='Stability', 
    y='Reward', 
    hue='Agent_Type',  # Using column names thanks to DataFrame integration
    style='Agent_Type', 
    s=100,             # Marker size
    palette='viridis'
)

plt.title('Agent Performance: Reward vs. Stability', fontsize=16, pad=20)
plt.xlabel('Stability Metric ($\sigma$)', fontsize=14)
plt.ylabel('Mean Cumulative Reward', fontsize=14)
plt.legend(title='Methodology', bbox_to_anchor=(1.05, 1), loc='upper left')

# 4. Output: Resolution
plt.tight_layout()
plt.savefig('agent_performance_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Plot generated and saved with high resolution.")
````
