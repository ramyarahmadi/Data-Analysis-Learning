 ستونه میشه

دو بعدی که سه سطر و دو ستون دارد

```python
a=np.array([[1,3],[4,6]])
print(a)
print(a.shape)
```

 دو بعدی که دو سطر  و دو ستون دارد

numpy ساخت آرایه سه بعدی بعدی (تنسور) توسط 

 سه بعدی چند سطر و چند ستون و چه تعداد میشه

```python
b=np.array([[[1,2,3]],[[-1,-3,0]],[[100,200,300]]])
print(b)
```

 سه بعدی که سه سطر و سه ستون که هر ستون یک عددی دارد

```python
print(b.shape)
```

```python
c=np.array([[[-1,-3,0],[1,2,3]],[[-1,-3,0],[-1,-3,0]],[[-1,-3,0],[100,200,300]]])
print(c)
```
