## In this lesson, we explored the following NumPy topics:

1. **dtype**: Determines the data type of an array’s elements.
2. **astype()**: Converts an array from one data type to another.
3. **Slicing**: Extracts specific portions of an array in one or more dimensions.
4. **reshape()**: Changes the shape of an array without modifying its data.
5. **flatten()**: Converts an array into a one-dimensional copy.
6. **concatenate()**: Combines multiple arrays into one.
7. **Masking**: Selects elements from an array using a Boolean condition.
8. **Random arrays**: Creates arrays with randomly generated values.

# Slicing

**Slicing 1-D Arrays(vector)**

```python
c=np.array([2,3,4,5,6,7])
c[2:4]    # Slicing in python means taking elements from one given index to another given index.
```

```python
c[:]      # whitout slicing 
```

```python
c[::]     # whitout slicing and the numbers of step!
```

**Negative Slicing :** Slice from the index 3 from the end to index 1 from the end

```python
print(c[-3:-1])    # -1 is the end of number in array
```

**STEP :** Use the step value to determine the step of the slicing

```python
print(c[1:5:2])   # number of step is 2
```

**Slicing 2-D Arrays(matrix)**

```python
d=np.array([[300,400,500,600],[3,4,5,6]])
print(f"The shape of d array is : {d.shape} index 0 is about the number of rows and another one (index 1) is the number columns ")
print(d)
```

```python
print('1',d[:])     # Select all elements from start to end
print('2',d[:, 1])  # Slicing: Get all rows from the second column (index 1)
print('3',d[1, :])  # Slicing: Get the second row (index 1) and all its columns
```

```python
print(d[0,:])      # Slicing: Get the first row (index 0) and all its columns
```

```python
print(d[0:2, 2])   # Select rows 0 and 1, and only the column at index 2
```

# reshape()

**Requirement for reshape()**: The product of the new dimensions must equal the total number of elements in the original array.

For an array with shape (2, 4), the product is 8. Valid reshapes include (8, 1), (4, 2), or (2, 2, 2).



**Old Shape Total Elements=New Shape Total Elements**

```python
a.reshape(3,1)   # 1x3 = 3x1 
```

```python
#d.shape was (2,4)  
print('1',d.reshape(4,2))
print('2',d.reshape(1,8))
print('3',d.reshape(8,1))
print('4',d.reshape(2,2,2))
```

**reshape(-1)** : Flatten the array into a 1D array

```python
print(d.reshape(-1))    
print(a.reshape(-1))
print(b.reshape(-1))
```

# flatten()

```python
print(a.flatten())
print(a.ravel())
print(a.reshape(-1))
```

# concatenate()

```python
d=np.array([[300,400,500,600],[3,4,5,6]])   #shape = (2,4)
e=np.array([[-1,0,1,2],[23,99,64,39]])      #shape = (2,4)
f=np.concatenate((d,e),axis=1)
print(f.shape)
f
```
