**ones :** Return a new array of given shape and type, filled with ones.

**Color Representation in NumPy Arrays**
When creating an array of shape `(10, 10, 3)`, we are defining an image with 10x10 pixels and 3 color channels (RGB).

By default, `np.ones` creates an array of type `float64`. In many visualization libraries like **Matplotlib**, floating-point color values are expected to be in the range $[0, 1]$:
- $0.0$ represents **Black**.
- $1.0$ represents **White**.

Since all values in `np.ones((10, 10, 3))` are exactly $1.0$, the resulting plot will appear as a solid **white** image.


```python
x = np.ones((4,4,3))     # The out put is float number 1 thats means the maximum between 0 to 1 mean whithe color 
x
```

```python
plt.imshow(x,cmap='gray')
```

**The Impact of Data Types on Color Representation**

It is crucial to understand how the `dtype` affects the visual output. Let's compare two scenarios:

1. **Floating-point (`float`):**
   `np.ones((10, 10, 3))` $\rightarrow$ Values are $1.0$. In the $[0, 1]$ scale, $1.0$ is **White**.

2. **Unsigned Integer (`uint8`):**
   `np.ones((10, 10, 3), dtype=np.uint8)` $\rightarrow$ Values are $1$. In the $[0, 255]$ scale, $1$ is nearly **Black**.

**Why does this happen?**
When using `uint8`, the computer expects values up to $255$ to represent full intensity. A value of $1$ is only $\approx 0.4\%$ of the maximum intensity, making it indistinguishable from black to the human eye. To get a **White** image with `uint8`, you must use `np.full((10, 10, 3), 255, dtype=np.uint8)`.

```python
x= np.ones((4,4,3),np.uint8)
x
```

```python
plt.imshow(x,cmap='gray')
```

**Creating White Color with `uint8`**

To represent a solid **White** color using the `uint8` data type, we cannot simply use `np.ones()`, as that would result in values of $1$ (which is nearly black). 

Instead, we use:
`x = np.ones((10, 10, 3), dtype=np.uint8) * 255`

**Breakdown of the operation:**
1. `np.ones((10, 10, 3), dtype=np.uint8)` $\rightarrow$ Creates an array where every element is **1**.
2. `* 255` $\rightarrow$ Performs element-wise multiplication, scaling every value from $1$ to **255**.
3. **Result:** Since $255$ is the maximum value for an 8-bit integer, the image is displayed as pure **White**.

```python
x = np.ones((4,4,3),np.uint8)*255
x
```

```python
plt.imshow(x,cmap='gray')
```

**Choose :** Construct an array from an index array and a list of arrays to choose from.

```python
import numpy as np
choices = [[0, 1, 2, 3], [10, 11, 12, 13],
  [20, 21, 22, 23], [30, 31, 32, 33]]
print(np.choose([2, 3, 1, 0], choices
# the first element of the result will be the first element of the
# third (2+1) "array" in choices, namely, 20; the second element
# will be the second element of the fourth (3+1) choice array, i.e.,
# 31, etc.
))
print(np.choose([2, 4, 1, 0], choices, mode='clip') # 4 goes to 3 (4-1)
# because there are 4 choice arrays
     )
np.choose([2, 4, 1, 0], choices, mode='wrap') # 4 goes to (4 mod 4)
# i.e., 0
```

## 3.2 Set routines

**union1d :** Return the unique, sorted array of values that are in either of the two input arrays.

```python
a = np.array([[4,5],[8,2]])
b = np.array([[2,5],[4,1]])
x = np.union1d(a,b)
x
```

**intersect1d :**  Return the sorted, unique values that are in both of the input arrays.

```python
a = np.array([[4,5],[8,2]])
b = np.array([[2,5],[4,1]])
x = np.intersect1d(a,b)
x
```

**unique :** Find the unique elements of an array.

```python
a = np.array([[1, 0, 0], [1, 0, 0], [2, 3, 4]])
np.unique(a, axis=0)
```

**setdiff1d :** Return the unique values in ar1 that are not in ar2.

```python
import numpy as np
a = np.array([1, 2, 3, 2, 4, 1])
b = np.array([3, 4, 5, 6])
np.setdiff1d(a, b)
```

## 3.3 Polynomials 

**polyval :** Evaluate a polynomial at specific values.

If p is of length N, this function returns the value:

p[0]*x**(N-1) + p[1]*x**(N-2) + ... + p[N-2]*x + p[N-1]

```python
a = np.array([2,3])       # dgree 1
x = np.polyval(a,2)
x  
```

```python
a = np.array([3,5,3])     # dgree 2           x=1    in 3x^2+5x+3
x = np.polyval(a,1)
x     
```

```python
a = np.array([3,0,3])     # dgree 2           x=1     in 3x^2+3
x = np.polyval(a,1)
x     
```

**polyder :**   (p, m=1)

Return the derivative of the specified order of a polynomial.

https://numpy.org/doc/2.1/reference/generated/numpy.polyder.html

```python
a = np.array([2,5,3])     # Derivative         in     2x^2+5x+3
x = np.polyder(a)
x   
```

**polydiv :** Returns the quotient and remainder of polynomial division.

This forms part of the old polynomial API. Since version 1.4, the new polynomial API defined in numpy.polynomial is preferred. A summary of the differences can be found in the transition guide.

```python
x = np.array([3.0, 5.0, 2.0])
y = np.array([2.0, 1.0])
np.polydiv(x, y)
```

## 5. Statistical functions

**mean :** Compute the arithmetic mean along the specified axis.

Returns the average of the array elements. The average is taken over the flattened array by default, otherwise over the specified axis. float64 intermediate and return values are used for integer inputs.

```python
a = np.array([5,2,4,1,3])
x = np.mean(a)
x
```

**median :**  Compute the median along the specified axis.

```python
a = np.array([2,4,1,3])
x = np.median(a)
x
```

**var :** Compute the variance along the specified axis.

Returns the variance of the array elements, a measure of the spread of a distribution. The variance is computed for the flattened array by default, otherwise over the specified axis.

```python
a = np.array([10,20])
x = np.var(a)
x
```

**std :** Compute the standard deviation along the specified axis.

Returns the standard deviation, a measure of the spread of a distribution, of the array elements. The standard deviation is computed for the flattened array by default, otherwise over the specified axis.

```python
a = np.array([10,20])
x = np.std(a)
x              
```
