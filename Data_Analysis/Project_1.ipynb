**This is a coin flip problem:** 

```python
faces=['H','T']
```

```python
import random
random.choice(faces)
```

```python
coins = random.choices(faces,k=20)
print(coins)
coins.count('H')
```

```python
NUM_SIMS = 100
NUM_FLIPS = 20
for sim in range(NUM_SIMS):
    coins = random.choices(faces, k = NUM_FLIPS)
    print(coins.count('H'), end=' ')
```

```python
import matplotlib.pyplot as plt
NUM_SIMS = 100
FLIPS = 20
results = []
for sim in range(NUM_SIMS):
    coins = random.choices(faces, k = FLIPS)
    #print(coins.count('H'), end=' ')
    results += [coins.count('H')]
```

```python
%matplotlib inline
```

```python
plt.scatter(range(NUM_SIMS),results);
plt.xlabel('Experiment number')
plt.ylabel('Number of heads observed')
plt.title('Simulation of no. of heads observed on 20 coin flips')
```

```python
random.seed(2882)
faces=['H','T']
num_sims=100
flips=20

results=[]
for sim in range(num_sims):              # The simulation loop
    coins=random.choices(faces, k=flips) # Simulate all coin flips for one experiment
    results+=[coins.count('H')]          # Add the number of 'H's in this experiment to our results list


plt.hist(results)
```
